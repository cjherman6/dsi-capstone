```python
kaggle_breeds = df['breed'].unique()
kaggle_breeds = [breed.lower() for breed in kaggle_breeds]
kaggle_breeds = [breed.replace(' ','-').replace('_','-') for breed in kaggle_breeds]
kaggle_breeds = sorted(kaggle_breeds)
kaggle_breeds
```

```python
url = 'https://dogtime.com/dog-breeds'
content = requests.get(url).content
soup = BeautifulSoup(content,'html.parser')
#soup
```

```python
content = soup.find_all(class_='post-title')
content
```

```python
dog_time_breeds = [((a.text.strip().replace('$',''))) for a in content]
dog_time_breeds = [breed.lower() for breed in dog_time_breeds]
dog_time_breeds = [breed.replace(' ','-').replace('_','-') for breed in dog_time_breeds]
dog_time_breeds = sorted(dog_time_breeds)
dog_time_breeds
```

```python
len(dog_time_breeds)
```

```python
# If the dog time breed is exactly in kaggle labels, add to suffix

scrape_suffix = []
kaggle_suffix = []
scrape_to_kaggle_dict = {}
remains = []

for dog in dog_time_breeds:
    if dog in kaggle_breeds:
        scrape_suffix.append(dog)
        kaggle_suffix.append(dog)
        scrape_to_kaggle_dict[dog] = dog
    else:
        remains.append(dog)

print(len(remains),len(scrape_suffix),len(kaggle_suffix))
# Of the dogs that didn't match kaggles labels, find those that
# contain the same words

scrape_suffix2 = []
kaggle_suffix2 = []

for dog in remains:
    for breed in kaggle_breeds:
        if breed in dog:
            scrape_suffix2.append(dog)
            kaggle_suffix2.append(breed)
            scrape_to_kaggle_dict[dog] = breed


print(len(remains),len(scrape_suffix2),len(kaggle_suffix2))
```

```python
len(kaggle_suffix + kaggle_suffix2)
```

```python
remainders = []

for dog in kaggle_breeds:
    if dog not in (kaggle_suffix + kaggle_suffix2):
        remainders.append(dog)
    else:
        pass
len(remainders)
```
